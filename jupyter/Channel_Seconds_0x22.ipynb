```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
pd.set_option('display.max_colwidth', None)
df = pd.read_csv(r"/mdfsvc/fep/log/Smart/OSMT_0x22-4.csv", lineterminator='\n', index_col='Date')
```

```python
"""
타임스탬프 변환 파트
"""
df["Datetime"] = df["Seconds"].apply(lambda x: datetime.fromtimestamp(x))
df["Timestamp(Convert)"] = df["Datetime"].apply(lambda x: x.strftime('%H:%M:%S'))
df = df[['Time', 'Message Type', 'Protocol ID', 'Timestamp(Convert)', 'Channel Index', 'Seconds', 'Raw Data']]
df
```
